3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        raise ValueError("The outcome x must be 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("The probability p must be between 0 and 1.")

    return p if x == 1 else (1 - p)

p = 0.7
outcomes = [0, 1]

print("Bernoulli PDF:")
for x in outcomes:
    print(f"P(X = {x}) = {bernoulli_pdf(x, p):.2f}")
```

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

def central_limit_theorem(n_samples, sample_size, distribution_func, *params):

    sample_means = []

    for _ in range(n_samples):
        sample = distribution_func(*params, size=sample_size)
        sample_means.append(np.mean(sample))

    return np.array(sample_means)

n_samples = 1000
sample_size = 30
distribution_func = np.random.exponential
lambda_param = 1

sample_means = central_limit_theorem(n_samples, sample_size, distribution_func, lambda_param)

plt.hist(sample_means, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black')
plt.title(f'Central Limit Theorem for Exponential Distribution (λ={lambda_param})')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

11. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.

```python
import numpy as np
import matplotlib.pyplot as plt

def verify_clt_normal(n_samples, sample_size, mean, std_dev):

    sample_means = []

    for _ in range(n_samples):
        sample = np.random.normal(mean, std_dev, sample_size)
        sample_means.append(np.mean(sample))

    return np.array(sample_means)

n_samples = 1000
sample_size = 30
mean = 0
std_dev = 1

sample_means = verify_clt_normal(n_samples, sample_size, mean, std_dev)

x = np.linspace(-4, 4, 1000)
population_distribution = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black', label='Sample Means')
plt.plot(x, population_distribution, label='Population Distribution (Normal)', color='red', linewidth=2)
plt.title('Central Limit Theorem Verification (Normal Distribution)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

    if p_value < alpha:
        result = "Reject the null hypothesis (significant difference)"
    else:
        result = "Fail to reject the null hypothesis (no significant difference)"

    return z_statistic, p_value, result

```

```python

sample_data = np.array([52, 55, 58, 50, 49, 51, 54, 56, 57, 55])
population_mean = 50
population_std = 5

z_statistic, p_value, result = z_test(sample_data, population_mean, population_std)

print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
print(result)
```

16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence_level=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)

    standard_error = sample_std / np.sqrt(sample_size)
    z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

    margin_of_error = z_value * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

```

```python
data = np.array([52, 55, 58, 50, 49, 51, 54, 56, 57, 55])
confidence_level = 0.95
lower, upper = confidence_interval(data, confidence_level)

print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)
    return data_skewness, data_kurtosis

mean = 0
std_dev = 1
size = 1000
data = np.random.normal(mean, std_dev, size)
skewness, kurt = calculate_skewness_kurtosis(data)

print(f"Skewness of the dataset: {skewness}")
print(f"Kurtosis (excess) of the dataset: {kurt}")

```
